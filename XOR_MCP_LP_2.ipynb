```python
import numpy as np
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)
np.random.seed(1)

W1 = np.random.rand(2,2)
b1 = np.random.rand(1,2)

W2 = np.random.rand(2,1)
b2 = np.random.rand(1,1)
lr=0.1
for i in range(10000):

    # Forward propagation
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    predicted_output = sigmoid(final_input)

    # Squared error (for loss only)
    loss = 0.5 * (y - predicted_output)**2

    # Backpropagation
    error = y - predicted_output
    d_predicted_output = lr * error * sigmoid_derivative(predicted_output)

    error_hidden_layer = d_predicted_output.dot(W2.T)
    d_hidden_layer = lr * error_hidden_layer * sigmoid_derivative(hidden_output)

    # Update weights
    W2 += hidden_output.T.dot(d_predicted_output)
    W1 += X.T.dot(d_hidden_layer)
print("Predicted Output:")
print(predicted_output)

```
